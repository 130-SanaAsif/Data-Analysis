```python
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

```python
flights.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=50)
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
#kde stands for kernel density estimation
```

```python
sns.pairplot(tips,hue='sex',palette = 'coolwarm')
```

```python
sns.rugplot(tips['tip'])
```

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Plot

```python
#barplot.
#it shows us the average or mean value of the data in x and y axis like in this example.
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips , hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex', split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

# Regression Plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,height=8)
```
