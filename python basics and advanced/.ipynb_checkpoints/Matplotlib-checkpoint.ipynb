```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional Method.
plt.plot(x,y)
#plt.show() -> Is only used when we are working some other editor rather than jupyter notebook.
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Graph")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'v--')

plt.subplot(1,2,2)
plt.plot(y,x,'y*-')
```

```python
#Object Oriented Method.
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b*--')
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
a = plt.figure()
axes1 = a.add_axes([0.1,0.1,0.8,0.8])
axes2 = a.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'g*-')
axes1.set_title("Larger Graph")
axes2.plot(y,x,'v--')
axes2.set_title("Smaller Graph")
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 3)
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,4,2,3],color='m', marker='o', linestyle='dashed', linewidth=2, markersize=12)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)

axes[0].plot(x,y,r)
axes[0].set_title("First Plot")

axes[1].plot(y,x,b)
axes[1].set_title("Second Plot")

axes[2].plot(x,y,g)
axes[2].set_title("Third Plot")

plt.tight_layout()
```
