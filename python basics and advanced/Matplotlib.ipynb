```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional Method.
plt.plot(x,y)
#plt.show() -> Is only used when we are working some other editor rather than jupyter notebook.
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Graph")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'v--')

plt.subplot(1,2,2)
plt.plot(y,x,'y*-')
```

```python
#Object Oriented Method.
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b*--')
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
a = plt.figure()
axes1 = a.add_axes([0.1,0.1,0.8,0.8])
axes2 = a.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'g*-')
axes1.set_title("Larger Graph")
axes2.plot(y,x,'v--')
axes2.set_title("Smaller Graph")
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 3)
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,4,2,3],color='m', marker='o', linestyle='dashed', linewidth=2, markersize=12)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)

axes[0].plot(x,y,'r')
axes[0].set_title("First Plot")

axes[1].plot(y,x,'b')
axes[1].set_title("Second Plot")

axes[2].plot(x,y,'g')
axes[2].set_title("Third Plot")

plt.tight_layout()
```

```python
fig = plt.figure()
axe = fig.add_axes([0,0,1,1])
axe.plot(x,x**2,label='X Square')
axe.plot(x,x**3,label='X Cube')

axe.legend(loc=(0.1,0.1))
```

# Lines and Marker Style

```python
fig,ax = plt.subplots(figsize = (18,9))

ax.plot(x, x+1, color = 'red', linewidth = 0.25)
ax.plot(x, x+2, color = 'red', linewidth = 0.50)
ax.plot(x, x+3, color = 'red', linewidth = 1.00)
ax.plot(x, x+4, color = 'red', linewidth = 2.00)

ax.plot(x, x+5, color = 'purple', lw = 3, linestyle = '-')
ax.plot(x, x+6, color = 'green', lw = 3, linestyle = '-.')
ax.plot(x, x+7, color = 'purple', lw = 3, linestyle = ':')

#custom dash
line, = ax.plot(x,x+8, color = 'black', lw=4)
line.set_dashes([5,10,15,10])

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+9,color='blue',lw=4,ls='-',marker='+')
ax.plot(x, x+10,color='green',lw=4,ls='--',marker='o')
ax.plot(x, x+11,color='yellow',lw=4,ls='--',marker='*')
ax.plot(x, x+12,color='magenta',lw=4,ls=':',marker=',')
ax.plot(x, x+13,color='red',lw=4,ls='-.',marker='.')
ax.plot(x, x+14,color='pink',lw=4,ls='--',marker='1')
ax.plot(x, x+15,color='orange',lw=4,ls='-.',marker='2')

# marker size and color
ax.plot(x, x+16, color = 'green', lw=3,ls='--',marker='o',markersize = 2)
ax.plot(x, x+17,color = 'yellow', lw=3,ls='-.',marker='+',markersize=3)
ax.plot(x,x+18,color = 'purple',lw=1,ls='-',marker='o',markersize=4,markerfacecolor='green')
ax.plot(x, x+19,color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");

```

# Special plot type

There are many specialized plots we can create, such as barplots, histograms, scatter plots, and much more. Most of these type of plots we will actually create using seaborn, a statistical plotting library for Python. But here are a few examples of these type of plots: 

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
data = [np.random.normal(0,std,100) for std in range (1,4)]
#rectangular box plot
plt.boxplot(data,vert = True, patch_artist=True)
```
