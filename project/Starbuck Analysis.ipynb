# Data visualization

```python
plt.figure(facecolor = "orange",figsize=(12,7))
sns.countplot(data=data,x='type')
plt.title("Types OF Procuct ")
```

```python
sns.catplot(data=data,x='protein',y='type',kind='bar')
plt.title("Protein in item type")
```

# Top 10 calories product

```python
cal = data[data['calories'] >=460]
```

```python
plt.figure(facecolor = 'orange',figsize=(12,7))
sns.countplot(data=cal,x='calories',hue='item')
plt.title(('Top 10 product of calories'))
```

```python
hical = cal['item'].count()
print(f"In the store there are {hical} products that have most calories")
```

```python
itemlist = cal['item'].tolist()
print('Top 10 calories items namelist:\n')
for item in itemlist:
    print(item)
```

```python
pro = data[data['protein'] >= 19]
```

```python
hipro = pro['item'].count()
print(f"In the store there are {hipro} products that have most calories")
```

```python
prolist = pro['item'].tolist()
print('Top 10 protein items namelist:\n')
for item in prolist:
    print(item)
```
